```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

HTML_PARSER = "html.parser"
ROOT_URL = 'http://www.ipeen.com.tw'
Output_total = pd.DataFrame(columns=['name','cate', 'score', 'average cost', 'link', 'phone', 'addr'])
LIST_URL = 'http://www.ipeen.com.tw/search/taiwan/000/1-0-0-0/'

def get_shop_link_list(soup):
    shop_info = soup.find_all('article', attrs={'class': 'serItem'})

    OutputData = pd.DataFrame(columns=['name','cate', 'score', 'average cost', 'link', 'phone', 'addr'])

    print(len(OutputData.index))
    for i,shop in enumerate(shop_info):
        name = (shop.find('a').text)
        score = (shop.find('li', attrs={'class': 'score'}).find('img').get('alt'))
        score = score[-2:]
        cate_type = (shop.find('li', attrs={'class': 'cate'}).text.strip())

        costEmpty = shop.find('li', attrs={'class': 'costEmpty'}).text.strip()
        costEmpty =costEmpty[costEmpty.find('消')+1:costEmpty.find('元')]
        link = ROOT_URL + shop.find('a', attrs={'data-label': '店名'})['href']
        shop_link = requests.get(link)
        sl = BeautifulSoup(shop_link.content, HTML_PARSER)

        phone = (sl.find('p', attrs={'class': 'tel i'}).text.strip())
        addr = (sl.find('p', attrs={'class': 'addr i'}).text.strip())

        OutputData.loc[i] = [name, cate_type, score, costEmpty, link, phone, addr]
    #print(OutputData)
        
    return OutputData


def get_page_and_req(i, OutputData, url):
    list_req = requests.get(url)
    global Output_total
    # try first 10 pages
    if i < 10:
        if list_req.status_code == requests.codes.ok:
            soup = BeautifulSoup(list_req.content, HTML_PARSER)
            newd = get_shop_link_list(soup)
            Output_total = pd.concat([Output_total ,newd])
            
            # find if there is next page
            npon = soup.find_all('div', attrs={'class':'next_page_area'})
            if(not len(npon)==0) :
                next_url = npon[0].find('a', attrs={'class': 'ga_tracking'}).get('href')
                print("next page")
                get_page_and_req(i+1, OutputData, next_url)
    
    return Output_total


if __name__ == '__main__':
    OutputData = pd.DataFrame(columns=['name','cate', 'score', 'average cost', 'link', 'phone', 'addr'])
    OutputData = get_page_and_req(0, OutputData, LIST_URL)
    OutputData = OutputData.set_index('name')
```
